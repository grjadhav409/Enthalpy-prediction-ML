```python
! pip install 'mordred[full]'  # install with extra requires

!pip install rdkit-pypi

import pandas as pd

from multiprocessing import freeze_support

from rdkit import Chem

from mordred import Calculator, descriptors
```

```python
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/data2/Enthalpy of formation enthalpy of formation/hydrocarbans/df_enthalpy_HC_smiles.xlsx")
```

```python
df
```

```python
df1 = df.drop_duplicates()
```

```python
df1
```

```python
name = df1['isomeric_smiles'].tolist()
molecules =[]
for i in name:
  molecules.append(Chem.MolFromSmiles(i))
if __name__ == "__main__":
    freeze_support()
    # Create Calculator
    calc = Calculator(descriptors)
df2= pd.DataFrame(data=calc.pandas(molecules))
df2
```

```python
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
```

```python
extracted_col_name= df1["name"]
extracted_col_hfg= df1["hfg"]
df2.insert(0, "name", extracted_col_name)
df2 = df2.join(extracted_col_hfg)
df2
```

```python
df2.to_excel(r"/content/drive/MyDrive/Colab Notebooks/data2/Enthalpy of formation enthalpy of formation/hydrocarbans/df_withMordred.xlsx", sheet_name='df_withMordred.xlsx', index = False)
```
